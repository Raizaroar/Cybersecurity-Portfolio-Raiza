```python
# Imports
import os
from collections import Counter
```

```python
# Cell 2: Function to analyze logs
def analyze_log(file_path, keyword):
    """
    Analyzes a log file and counts occurrences of a keyword.
    
    Args:
        file_path (str): Path to the log file.
        keyword (str): Word to search for.
    
    Returns:
        tuple: (counter, lines_found)
    """
    try:
        with open(file_path, "r") as file:
            lines = file.readlines()
    
        counter = 0
        lines_found = []
        
        for line in lines:
            if keyword.lower() in line.lower():
                counter += 1
                lines_found.append(line.strip())
        
        return counter, lines_found
    
    except FileNotFoundError:
        print(f"Error: The file {file_path} does not exist")
        return 0, []
    except Exception as e:
        print(f"Unexpected error: {e}")
        return 0, []
```

```python
# Analyze test log file
path = ‘../data/sample_log.txt’
words = [‘error’, ‘failed’, ‘unauthorized’]

print(“=” * 60)
print(“ LOG FILE ANALYSIS”)
print(“=” * 60)

for word in words:
    count, lines = analyze_log(path, word)
    print(f“\n Search: ‘{word.upper()}’”)
    print(f“ Occurrences: {count}”)
    if lines:
    print(“Lines found:”)
    for line in lines:
        print(f“  → {line}”)
```
